```python
import matplotlib.pyplot as mat
import pandas as pd
import numpy as np
```

# Problem #1

```python
# Read in the data
students_df = pd.read_csv('student_class.csv')
median_values = students_df.groupby('Majors').median()
chart_data = median_values[['Midterm', 'Final', 'Final Average']]

# Get labels and calculate spacing between them from the selected dataset
labels = chart_data.index
label_locations = np.arange(len(labels))
width = 0.20

# Create bars and set a size for chart
mat.bar(x=label_locations - width, height=chart_data['Midterm'], width=width, label='Midterm', color='red')
mat.bar(x=label_locations, height=chart_data['Final'], width=width, label='Final', color='green')

mat.bar(x=label_locations + width, height=chart_data['Final Average'], width=width, label='Final Avg', color='blue')
mat.gcf().set_size_inches(16, 5)

# Details for Graph
mat.title('Median Midterm, Final, and Final Average Scores')
mat.xlabel('Major')
mat.ylabel('Median Score')

mat.xticks(label_locations, labels=labels)
mat.legend(loc='upper right') 

mat.show()
```

Chart Info.

This visualization represents median test scores for a midterm, final, and final average for some course. Marks for this graph are rectangular bars which represent the indivdual grade and are colored in red, green, and blue (channels). Attributes of this graph include the labels (median score and major), the title, and the legend.

Questions

Upon first look, BNFO majors seem to be doing the best in terms of midterm and final grades, although there seems to be an even score amongst all majors for final average.

# Problem #2

```python
# Set data for histogram
bin_data = students_df['Final Average']
bins = np.sort(np.append(students_df['Numeric Bin'].unique(), 100))

# Plot histogram
mat.hist(bin_data, bins=bins)
mat.gcf().set_size_inches(18, 6)

# Set title and axes names
mat.title('Number of Students in Each Grade Bin')
mat.xlabel('Grade')
mat.ylabel('Frequency')

mat.show()
```

Chart Info.

This visualization represents amount of students with a certain grade (represented in bins). Marks for this graph are physical rectangular bars (or bins) which are spread amongst a certain grade bin, and measured for frequency. Channels for this graph are the blue color as well as the shape of the histogram bars. Attributes of this graph include the labels (frequency and grade) and the title.
